```python
import cv2
import matplotlib.pyplot as plt

# template matching
img = cv2.imread("/content/drive/MyDrive/template_matching/6.jpg", 0)
print(img.shape)

template = cv2.imread("/content/drive/MyDrive/template_matching/2.1.jpg", 0)
print(template.shape)

h, w = template.shape

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for meth in methods:
    method = eval(meth) # 'cv2.TM_CCOEFF' -> cv2.TM_CCOEFF

    res = cv2.matchTemplate(img, template, method)
    print(res.shape)
    min_val, max_val, min_loc, max_loc=cv2.minMaxLoc(res)

    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc

    bottom_right = (top_left[0] + w, top_left[1] + h)

    cv2.rectangle(img, top_left, bottom_right, 255, 2)

    plt.figure()
    plt.subplot(121), plt.imshow(res, cmap = "gray")
    plt.title("Eşleşen Sonuç"), plt.axis("off")
    plt.subplot(122), plt.imshow(img, cmap = "gray")
    plt.title("Tespit edilen Sonuç"), plt.axis("off")
    plt.suptitle(meth)

    plt.show()



```

```python
from google.colab.patches import cv2_imshow
import cv2
import os

def en_cok_benzeyen_resmi_bul(klasor_yolu, template_yolu):
    # Template'i okuyun
    template = cv2.imread(template_yolu, 0)

    # Klasördeki tüm resim dosyalarını alın
    dosya_listesi = os.listdir(klasor_yolu)

    en_yuksek_benzerlik_skoru = 0
    en_cok_benzeyen_resim = None

    for dosya_adi in dosya_listesi:
        # Dosya yolunu oluşturun
        dosya_yolu = os.path.join(klasor_yolu, dosya_adi)

        # Sadece resim dosyalarını işleyin
        if dosya_adi.endswith(".jpg") or dosya_adi.endswith(".jpeg") or dosya_adi.endswith(".png"):
            # Resmi okuyun
            resim = cv2.imread(dosya_yolu, 0)

            # Template ve resim boyutlarını eşitleyin
            template = cv2.resize(template, (resim.shape[1], resim.shape[0]))

            # Desen eşleştirmeyi gerçekleştirin
            benzerlik_skoru = desen_eslestirme(template, resim)

            # En yüksek benzerlik skorunu güncelleyin
            if benzerlik_skoru > en_yuksek_benzerlik_skoru:
                en_yuksek_benzerlik_skoru = benzerlik_skoru
                en_cok_benzeyen_resim = resim

    return en_cok_benzeyen_resim

def desen_eslestirme(template, resim):
    # Desen eşleştirmeyi gerçekleştirin
    res = cv2.matchTemplate(resim, template, cv2.TM_CCOEFF_NORMED)

    # Eşleşme skorlarını alın
    benzerlik_skorlari = cv2.minMaxLoc(res)[1]

    # En yüksek benzerlik skorunu döndürün
    return benzerlik_skorlari

# Klasör ve template yollarını belirleyin
klasor_yolu = "/content/drive/MyDrive/template_matching"
template_yolu = "/content/drive/MyDrive/2.2.jpg"

# En çok benzeyen resmi bulun
en_cok_benzeyen_resim = en_cok_benzeyen_resmi_bul(klasor_yolu, template_yolu)

# En çok benzeyen resmi görüntüleyin
cv2_imshow(en_cok_benzeyen_resim)
```
